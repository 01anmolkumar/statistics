```python
set1={34,67,78,100}
set2={67,78}
```

```python
#difference
set1-set2
```

```python
#subset
set2.issubset(set1)
```

```python
#symetric difference
set1={34,67,78,100}
set2={67,78,102}
set1.symmetric_difference(set2)
```
